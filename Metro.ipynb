### 2.5.2 Further clean up column names

```python
"""I have a dataframe (distress_df) with columns that have a uuid at the end that I want to get rid of see the example column listing below. write a function to shorten to a meaningful column name:
## columns
Index(['Distress:Date Picker Label(15-1600C5B3-6A98-4E89-BC45-DA4426D0C24C-1649252806239)',
       'Distress:Chain marker (xxx.xx)(15-1600C5B3-6A98-4E89-BC45-DA4426D0C24C-1649252511945)',
       'Distress:Segment Rating(15-1600C5B3-6A98-4E89-BC45-DA4426D0C24C-1651863175838)',
       'Distress:Comments(15-1600C5B3-6A98-4E89-BC45-DA4426D0C24C-1652128331926)',
       'Distress:Fill out form below IF RATING CHANGES FROM ABOVE(518-9BA0A758-F7E6-4B9E-A1C3-B516BE8271C9-1698786868844)',
       'Distress:2023 Inspection Date(518-9BA0A758-F7E6-4B9E-A1C3-B516BE8271C9-1698786546348)',
       'Distress:2023 Segment Rating(518-9BA0A758-F7E6-4B9E-A1C3-B516BE8271C9-1698786588363)',
       'Distress:2023 Inspection Comments(518-9BA0A758-F7E6-4B9E-A1C3-B516BE8271C9-1698786826879)'],
      dtype='object')"""

# Solution
import re

def clean_column_names(df):
    # Define a pattern to extract the main part of the column name before the UUID
    pattern = r"(Distress:[\w\s]+)"

    # Use the regex pattern to extract the clean name and remove any trailing spaces
    new_columns = [re.search(pattern, col).group(1).strip() if re.search(pattern, col) else col for col in df.columns]

    # Rename the columns of the DataFrame
    df.columns = new_columns
    return df

# Example usage:
# Assuming distress_df is your DataFrame with the original long column names
clean_distress_df = clean_column_names(distress_df)
print(clean_distress_df.columns)
```

Using Matplotlib library

```python
import numpy as np
# Filter the DataFrame to include only rows where SegmentRatingInt is 2 or greater
df = rating_df[rating_df['SegmentRatingInt'] >= 2]

# Then, round ChainMarker to whole number
df['ChainMarker'] = df['ChainMarker'].apply(np.floor)

# Finally, group by ChainMarker and count the number of rows
grouped_df = df.groupby('ChainMarker').size()

# Set gap width
gap = 0.1

# Create a new figure
fig, ax = plt.subplots()

# Create bar plot
ax.bar(grouped_df.index, grouped_df.values, width=1-gap)

ax.set_xlabel('ChainMarker +00 (ft)')
ax.set_ylabel('Count of 10 ft SegmentRatingInt >= 2')
ax.set_title('Number of Wet Rated Segements per 100-ft, B-Line Track 1')

# Compute the running sum
running_sum = grouped_df.cumsum()

# Create a secondary y-axis
ax2 = ax.twinx()

# Plot the running sum on the secondary y-axis
ax2.plot(running_sum.index, running_sum.values, color='r', marker='o')

ax2.set_ylabel('Running Sum')

ax.legend(loc='upper left')
ax2.legend(loc='upper right')
ax.set_ylim([0, 10])
ax2.set_ylim([0, 200])

plt.tight_layout()
plt.show()
```
